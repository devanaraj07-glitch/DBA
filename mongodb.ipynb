```python
orders = pd.read_csv(base_path + "orders.csv")
deliveries = pd.read_csv(base_path + "deliveries.csv")
customers = pd.read_csv(base_path + "customers.csv")
complaints = pd.read_csv(base_path + "complaints.csv")
incidents = pd.read_csv(base_path + "incidents.csv")
app_events = pd.read_csv(base_path + "app_events.csv")
```

Inserting Customers (with embedded complaints)

```python
customer_docs = []

for _, c in customers.iterrows():
    cust_complaints = complaints[complaints["customer_id"] == c["customer_id"]]

    complaint_list = []
    for _, comp in cust_complaints.iterrows():
        complaint_list.append({
            "complaint_id": comp["complaint_id"],
            "complaint_type": comp["complaint_type"],
            "severity": comp["severity"],
            "status": comp["status"],
            "resolution_days": comp["resolution_days"]
        })

    customer_docs.append({
        "customer_id": c["customer_id"],
        "age": c["age"],
        "home_zone": c["home_zone"],
        "customer_type": c["customer_type"],
        "signup_date": c["signup_date"],
        "loyalty_score": c["loyalty_score"],
        "app_engagement_score": c["app_engagement_score"],
        "preferred_channel": c["preferred_channel"],
        "account_status": c["account_status"],
        "complaints": complaint_list
    })

db.customers.insert_many(customer_docs)
```

Inserting Deliveries (with embedded incidents)

```python
delivery_docs = []

for _, d in deliveries.iterrows():
    related_incidents = incidents[incidents["delivery_id"] == d["delivery_id"]]

    incident_list = []
    for _, i in related_incidents.iterrows():
        incident_list.append({
            "incident_type": i["incident_type"],
            "severity": i["severity"],
            "reported_at": i["reported_at"],
            "resolution_status": i["resolution_status"],
            "resolved_hours": i["resolved_hours"]
        })

    delivery_docs.append({
        "delivery_id": d["delivery_id"],
        "order_id": d["order_id"],
        "driver_id": d["driver_id"],
        "vehicle_id": d["vehicle_id"],
        "hub_id": d["hub_id"],
        "dispatch_time": d["dispatch_time"],
        "delivery_completed_at": d["delivery_completed_at"],
        "delivery_status": d["delivery_status"],
        "route_distance_km": d["route_distance_km"],
        "manual_route_override_count": d["manual_route_override_count"],
        "proof_of_completion_missing": d["proof_of_completion_missing"],
        "customer_rating_post_delivery": d["customer_rating_post_delivery"],
        "fuel_or_charge_cost": d["fuel_or_charge_cost"],
        "incidents": incident_list
    })

db.deliveries.insert_many(delivery_docs)
```
